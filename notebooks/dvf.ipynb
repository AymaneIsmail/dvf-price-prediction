```python
import pandas as pd
```

```python
DVF_FILE = '../data/raw/dvf.csv'
chunk_size = 100_000 
```

```python
null_counts = None

for chunk in pd.read_csv(DVF_FILE, chunksize=chunk_size):
    if null_counts is None:
        null_counts = chunk.isnull().sum()
    else:
        null_counts += chunk.isnull().sum()

print(null_counts)
```

```python
total_rows = 0
null_counts = None

for chunk in pd.read_csv(DVF_FILE, chunksize=chunk_size):
    total_rows += chunk.shape[0]
    if null_counts is None:
        null_counts = chunk.isnull().sum()
    else:
        null_counts += chunk.isnull().sum()

print("Valeurs nulles par colonne :\n", null_counts)
print("\nPourcentage par colonne :\n", (null_counts / total_rows) * 100)
```
